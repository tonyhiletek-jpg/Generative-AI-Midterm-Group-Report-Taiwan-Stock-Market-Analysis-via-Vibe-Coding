```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# 1. 定義族群清單
stock_groups = {
    "AI 伺服器族群": ["2382.TW", "2317.TW", "3231.TW", "6669.TW", "2356.TW"], # 廣達、鴻海、緯創、緯穎、英業達
    "半導體權值族群": ["2330.TW", "3711.TW", "2303.TW", "2337.TW"], # 台積電、日月光、聯電、旺宏
    "IC 設計族群": ["2454.TW", "3661.TW", "3034.TW", "2379.TW", "8046.TW"]    # 聯發科、世芯-KY、聯詠、瑞昱、南電
}

def plot_sector_index_k_line(group_name, tickers, period="6mo"):
    all_data = []
    print(f"正在抓取並計算 {group_name} 數據...")

    for ticker in tickers:
        # VS Code 環境中建議 auto_adjust=True 避免欄位名稱混亂
        df = yf.download(ticker, period=period, auto_adjust=True, progress=False)
        
        # 處理 yfinance 可能回傳的 MultiIndex 欄位
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = df.columns.get_level_values(0)
            
        if not df.empty:
            all_data.append(df[['Open', 'High', 'Low', 'Close', 'Volume']])

    if not all_data:
        print(f"無法抓取 {group_name} 的數據。")
        return

    # 計算族群指數（取平均值）
    group_df = pd.concat(all_data).groupby(level=0).mean()
    # 總成交量則取總和
    group_df['Volume'] = pd.concat([d['Volume'] for d in all_data], axis=1).sum(axis=1)

    # 計算均線
    group_df['MA5'] = group_df['Close'].rolling(window=5).mean()
    group_df['MA20'] = group_df['Close'].rolling(window=20).mean()

    # 2. 建立繪圖
    fig = make_subplots(
        rows=2, cols=1, shared_xaxes=True,
        vertical_spacing=0.08,
        subplot_titles=(f'<b>{group_name} 指數走勢 (平均值)</b>', '<b>族群總成交量</b>'),
        row_heights=[0.75, 0.25]
    )

    # K線圖
    fig.add_trace(go.Candlestick(
        x=group_df.index, open=group_df['Open'], high=group_df['High'],
        low=group_df['Low'], close=group_df['Close'], name='族群指數',
        increasing_line_color='#ef5350', decreasing_line_color='#26a69a'
    ), row=1, col=1)

    # 均線
    fig.add_trace(go.Scatter(x=group_df.index, y=group_df['MA5'], line=dict(color='#2196F3', width=1.5), name='5MA'), row=1, col=1)
    fig.add_trace(go.Scatter(x=group_df.index, y=group_df['MA20'], line=dict(color='#FF9800', width=1.5), name='20MA'), row=1, col=1)

    # 成交量
    vol_colors = ['#ef5350' if group_df['Close'].iloc[i] >= group_df['Open'].iloc[i] else '#26a69a' for i in range(len(group_df))]
    fig.add_trace(go.Bar(x=group_df.index, y=group_df['Volume'], name='總成交量', marker_color=vol_colors), row=2, col=1)

    # 佈局設定
    fig.update_layout(
        title={'text': f"<b>{group_name} 趨勢判讀分析</b>", 'x':0.5, 'font':{'size':22}},
        xaxis_rangeslider_visible=False,
        height=800,
        template='plotly_white',
        hovermode='x unified'
    )

    # 排除週末空窗期
    fig.update_xaxes(rangebreaks=[dict(bounds=["sat", "mon"])])
    
    # 在 VS Code 中這會開啟瀏覽器顯示圖表
    fig.show()

# 3. 執行迴圈
if __name__ == "__main__":
    for group, tickers in stock_groups.items():
        plot_sector_index_k_line(group, tickers)
```
